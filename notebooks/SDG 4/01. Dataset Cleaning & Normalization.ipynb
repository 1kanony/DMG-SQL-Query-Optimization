```python
import pandas as pd
from pathlib import Path
```

#### Extract

```python
# Reading the dataset

DATASETS_DIR = Path('../../datasets/UNESCO SDG/SDG')

datasets_paths = {
    'country': DATASETS_DIR / 'SDG_COUNTRY.csv',
    'data_national': DATASETS_DIR / 'SDG_DATA_NATIONAL.csv',
    'data_regional': DATASETS_DIR / 'SDG_DATA_REGIONAL.csv',
    'label': DATASETS_DIR / 'SDG_LABEL.csv',
    'metadata': DATASETS_DIR / 'SDG_METADATA.csv',
    'region': DATASETS_DIR / 'SDG_REGION.csv',
}

datasets = {}

for k,v in datasets_paths.items():
    datasets[k] = pd.read_csv(v, low_memory=False)
```

#### Transform

##### Cleaning

###### SDG_COUNTRY.csv

```python
country_df = datasets['country']

country_df['COUNTRY_ID'] = country_df['COUNTRY_ID'].str.strip()
country_df['COUNTRY_NAME_EN'] = country_df['COUNTRY_NAME_EN'].str.strip()
country_df = country_df.drop_duplicates(subset=['COUNTRY_ID'])
```
